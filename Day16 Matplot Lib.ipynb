# Bar Graph

```python
x=[1,2,3]
y=[4,5,6]
c=['r','g','b']
plt.bar(x,y,color=c)
plt.show()
```

```python
import numpy as np
```

```python
x=[1,2,3]
y=[4,5,6]
c=['r','g','b']
plt.barh(x,y,color=c)
plt.show()
```

```python
x=np.array(["A","B","c"])
y=np.array([2,4,6])
plt.barh(x,y,color="red")
plt.show()
```

```python
x=np.array(["A","B","c"])
y=np.array([2,4,6])
plt.barh(x,y,color="green", height=0.2)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt 
```

```python
x=[1,3,5,7,9]
```

```python
y=[2,3,2,6,8]
```

```python
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel("numbers")
c=["red","magenta","yellow","green","blue"]
plt.scatter(x,y, color=c)
plt.show()
```

```python
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel("numbers")
c=["red","magenta","yellow","green","blue"]
plt.scatter(x,y, color=c, s=150)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
fname=r'boy.jpg' 
```

```python
#opening image using PIL 
image = Image.open(fname).convert('L')

#mapping image to grayscale 
plt.imshow(image, cmap='gray')
plt.show()
```

```python
#opening image using PIL 
image = Image.open(fname).convert('L')

#mapping image to grayscale 
plt.imshow(image, cmap='PiYG')
plt.show()
```

```python
#opening image using PIL 
image = Image.open(fname).convert('RGB')

#mapping image to original 
plt.imshow(image)
plt.show()

```

```python
#opening image using PIL 
image = Image.open(fname).convert('L')
#mapping image to grayscale or desired colour  
plt.title("Small Boy")
plt.xlabel("Length")
plt.ylabel("Breadth")
plt.imshow(image, cmap='gist_heat_r')
plt.grid()
plt.show()
```

# Matplotlib PIE

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['english','hindi','math','science']
plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=['english','hindi','math','science']
c=['red','magenta','blue','green']
plt.pie(x,labels=y, colors=c)
plt.legend()
plt.show()
```
